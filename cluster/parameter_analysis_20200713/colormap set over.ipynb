```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm
```

```python
a = np.random.random((100, 100))
a.sort()
```

```python
cmap = cm.get_cmap()
cmap.set_over("k")
```

```python
plt.imshow(a, cmap=cmap)
plt.colorbar()
```

```python
plt.imshow(a, vmax=0.5, cmap=cmap)
plt.colorbar(extend="max")
```
